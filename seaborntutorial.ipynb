```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
sns.get_dataset_names()
```

```python
tips =sns.load_dataset("tips")
print(tips)
```

```python
sns.set_theme(style ="ticks")
sns.relplot(
    data = tips,
    x ="total_bill",
    y = "tip",
    kind ="line",
    hue ="smoker",
    size ="size",
    style = "smoker"
)
```

```python
tips =sns.load_dataset("tips")
sns.barplot(
    data = tips,
    x ="day",
    y ="tip",
    hue ="sex"
)

```

```python
sns.boxplot(
    data = tips,
    x ="day",
    y ="tip",
    hue ="sex"
)
```

```python
pengu = sns.load_dataset("penguins")
sns.histplot(
    data =pengu,
    x ="body_mass_g",
    bins =30
)
```
