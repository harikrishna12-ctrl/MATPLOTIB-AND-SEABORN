```python
# Real data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]   # in $M

# Plot
plt.plot(years, oscar_revenue)

plt.title("Oscar Movies Revenue in each Year")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
```

```python
# Non-Oscar Winner Movies

non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_years = [2008, 2009, 2010, 2011, 2012]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $m

plt.plot(years, oscar_revenue,"o--r",label ="oscar movies")
plt.plot(years, non_oscar_revenue,"<--g",label = "Non oscar movie")

plt.title("oscar vs non_oscar movies")
plt.xlabel("year")
plt.ylabel("Revenue ( in $m)")

plt.legend()
```

```python
movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]   # in $M

plt.bar(years,oscar_revenue, color = "red")
plt.title("Revenue of movies in years")
plt.xlabel("years")
plt.ylabel("Revenue (in $m)")

plt.text(years[0],oscar_revenue[0] + 10,"The Dark Knight",ha = "center")

plt.ylim(0,max(oscar_revenue) + 100)
```

```python

movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]   # in $M

non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_years = [2008, 2009, 2010, 2011, 2012]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $m

x = np.arange(len(years))
width = 0.2
plt.bar((x - width)/2,oscar_revenue,width,label = "oscar movies")
plt.bar((x+width)/2,non_oscar_revenue,width,label = "non oscar movie")
plt.title("oscar vs non oscar")
plt.xlabel("years")
plt.ylabel("Revenue (in $m)")

plt.xticks(x,years)
```

```python
people = [
    "Person A", "Person B", "Person C", "Person D", "Person E",
    "Person F", "Person G", "Person H", "Person I", "Person J"
]

age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]

bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

plt.scatter(age,bp,cmap = "plasma",c= bp)
plt.title("AGE VS BP")
plt.xlabel("age")
plt.ylabel("Blood pressure")
plt.grid()
plt.colorbar()

for i in range(len(age)):
    plt.annotate(people[i], xy=(age[i],bp[i]), xytext = (age[i]+2,bp[i]+2))
plt.xlim(min(age), max(age)+10)
plt.ylim(min(bp),max(bp) +5)
```
